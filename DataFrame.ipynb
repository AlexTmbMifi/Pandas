```python
# данные о чаевых в некоторм кофе
data_df = pd.read_csv('tips.csv')
data_df
```

```python
# apply с несколькими переменными
# оцениваем щедрые или обычные чаевые
def quality(total_bill, tip):
    if tip/total_bill > 0.15:
        return 'Щедрые чаевые'
    else:
        return 'Обычные чаевые'
```

```python
quality(15, 5.05)
```

```python
# Inner join or merge
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
logins
```

```python
registrations
```

```python
# Внутреннее соединение по имени: в обоих фреймах есть имя Андрей и Бобо
pd.merge(registrations,logins,how='inner',on='name')
# здесь нет разницы, где registrations и logins
# кроме того, колонки с разными названиями , но одинаковой размерности можно объединять
# для этого существуют параметры left_on и right_on
```

```python
# Left and right join. Лево- и правостороннее слияние.
# Левостороннее: левая основная таблица. Если нет фактора, по которому можно соединенить две таблицы, то для столбцов правой таблицы пишется NaN
# Правостороннее соединение абсолютно аналогично
```

```python
pd.merge(left=registrations, right = logins, how='left', on='name')
```

```python
pd.merge(left=registrations, right = logins, how='right', on='name')
```

```python
# Outer merge или внешнее слияние.
# Мы берём все строки из двух таблиц
# если нет совпадений по on , то пишется NaN, но мы сохраненяем все данные двух таблиц.
```

```python
pd.merge(left=registrations, right = logins, how= 'outer', on='name')
```

**На какой тип вопросов помогает ответить pivot?**

**Представим себе такую задачу. Вы хотите узнать, сколько лицензий каждого продукта было куплено компанией Google. Сейчас данные отформатированы таким образом, что нам сложно увидеть ответ на этот вопрос. Давайте с помощью pivot поменяем данные - возьмём только те колонки, которые помогут нам ответить на наш вопрос.**

```python
# Возьмём поднабор данных - только отдельные колонки
licenses = df[['Company','Product','Licenses']]
licenses
```

```python
pd.pivot(data=licenses,index='Company',columns='Product',values='Licenses')
```

```python
# т.е. сводные таблицы позволяют поменять структуру данных, преобразуя ячейки данных с помощью нового индекса
# новый индекс в данных Company. В качестве столбцов теперь Product, а в качестве значений - лицензии.
# По итогу, нового ничего нет. Произошла переструктуризация данных
```

```python
pd.pivot_table(df,index="Company",aggfunc='sum')[['Licenses','Sale Price']]
```

```python
# Либо обозначаем их как целевые значения (values)
pd.pivot_table(df,index="Company",aggfunc='sum',values=['Licenses','Sale Price'])
```
