```python
# данные о чаевых в некоторм кофе
data_df = pd.read_csv('tips.csv')
data_df
```

```python
# apply с несколькими переменными
# оцениваем щедрые или обычные чаевые
def quality(total_bill, tip):
    if tip/total_bill > 0.15:
        return 'Щедрые чаевые'
    else:
        return 'Обычные чаевые'
```

```python
quality(15, 5.05)
```

```python
data_df[['total_bill', 'tip']].apply(lambda x: quality(x['total_bill'], x['tip']), axis=1) # важно указывать ось, потому что работаем со столбцами
```

```python
data_df['quality'] = np.vectorize(quality)(data_df['total_bill'], data_df['tip']) # работает производительнее и удобнее
# проводим викторизацию, чтобы функция на входе понимала, что к ней идут серии
```

```python
# Конкатенация - склеивание 
# можно делать по столбцам и по строкам
```

```python
# иницилизация словарей
data_one = {'A': ['A0', 'A1', 'A2', 'A3'],'B': ['B0', 'B1', 'B2', 'B3']}
data_two = {'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']}
```

```python
# инициализация фреймов
one = pd.DataFrame(data_one)
two = pd.DataFrame(data_two)
one
```

```python
# Конкатенация по строкам
axis0 = pd.concat([one,two],axis=0)
axis0
```

```python
# Конкатенация по столбцам, более понятная и читаемое склеивание
axis1 = pd.concat([one,two],axis=1)
axis1
```

```python
# Переименовываем колонки
two.columns = one.columns
```

```python
# более привычное склеивание по строкам
axis0 = pd.concat([one,two],axis=0)
axis0
```
