```python
import pandas as pd
```

```python
# группировка по отдельным категориям
df1 = pd.read_csv('mpg.csv')
df1
```

```python
df1['model_year'].value_counts() # здесь внутри зашита 
```

```python
df1.groupby('model_year').aggregate({'weight':'mean'}) # группирока данных + агрегация
```

```python
df1.groupby(['model_year','cylinders']).aggregate({'weight':'mean'}) # группирока данных по двум колонок + агрегация 
```

```python
df1.groupby(['model_year','cylinders']).aggregate({'weight':'mean'}).index # описание мультииндексов и их названия
```

```python
# можно получить различные метрики для таблицы
df1.groupby('model_year').describe()
```

```python
df1.groupby(['model_year','cylinders']).aggregate({'weight':'mean'}).index.names # названия индексов
```

```python
df1.groupby(['model_year','cylinders']).aggregate({'weight':'mean'}).index.levels # уровни
```

```python
df1.groupby(['model_year','cylinders']).aggregate({'weight':'mean'}).loc[70] # с 70 моделью работаем
```

```python
df1.groupby(['model_year','cylinders']).aggregate({'weight':'mean'}).loc[[70,82]] # с 70 и 82 моделью работаем
```

```python
df1.groupby(['model_year','cylinders']).aggregate({'weight':'mean'}).loc[(70,4)] # получение кокретного значения
```
